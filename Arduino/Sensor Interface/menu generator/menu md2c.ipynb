```python
def split(s):
    try:
        n = s.index(';')
        return s[:n], s[n+1:]
    except:
        return '', s
    
lines = open("menu.md", "r").readlines()
N = len(lines)
indention = [l.count('    ') for l in lines]
lines = [l.strip().upper().replace('- ', '') for l in lines]
lines = [""] + lines
lines = [split(l) for l in lines]
refs, lines = list(zip(*lines))
```

```python
next_list = [0 for i in range(N+1)]
previous_list = [0 for i in range(N+1)]
parent_list = [0 for i in range(N+1)]
child_list = [0 for i in range(N+1)]
```

```python
def append_child(p, c):
    parent_list[c] = p
    child_list[p] = c

def append_next(this, n):
    next_list[this] = n
    previous_list[n] = this
    parent_list[n] = parent_list[this]
```

```python
stack = []
for i in range(1, N):
    if indention[i] == indention[i-1]:
        append_next(i, i+1)
        continue
    
    if indention[i] > indention[i-1]:
        stack.append(i-1)
        append_child(i, i+1)
        
    if indention[i] < indention[i-1]:
        for j in range(indention[i-1] - indention[i] - 1):
            stack.pop()
        append_next(stack.pop()+1, i+1)
```

```python
for i in range(1, len(lines)):
    print(i, '  '*indention[i-1], lines[i])
```
